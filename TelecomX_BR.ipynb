#ðŸ“Œ ExtracÃ£o

```python
import pandas as pd
```

```python
# Carrega dados JSON
df = pd.read_json('TelecomX_Data.json')
df
```

```python
customer_normalizado = pd.json_normalize(df['customer'])
customer_normalizado
```

```python
phone_normalizado = pd.json_normalize(df['phone'])
phone_normalizado
```

```python
internet_normalizado = pd.json_normalize(df['internet'])
internet_normalizado
```

```python
account_normalizado = pd.json_normalize(df['account'])
account_normalizado
```

```python
# Unifica as vaariÃ¡veis em um dataframe
df = pd.concat([df[['customerID','Churn']], customer_normalizado, phone_normalizado, internet_normalizado, account_normalizado], axis=1)
df
```

#ðŸ”§ TransformaÃ§Ã£o

```python
(df == '').sum()
```

```python
# Substituindo os espaÃ§os em branco por NaN
df['Charges.Total'] = df['Charges.Total'].replace(' ', pd.NA)
df['Churn'] = df['Churn'].replace('', pd.NA)
```

```python
# Converte para numÃ©rica a coluna Charges.Total
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
```

```python
df[(df['StreamingMovies'] != 'Yes') & (df['StreamingMovies'] != 'No')]
```

```python
# Converte para bool
colunas_binarias = ['SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                    'StreamingTV', 'StreamingMovies', 'PaperlessBilling']

for coluna in colunas_binarias:
    df[coluna] = df[coluna].replace({'Yes': True, 'No': False, 0: False, 1: True, 'No phone service': False, 'No internet service': False})
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
# Removendo as linhas que contÃªm NaN
df.dropna(inplace=True)

# Remove colunas relevantes para a anÃ¡lise de evasÃ£o
df.drop(columns=['customerID'], inplace=True)
```

```python
# Converte a coluna target para bool
df['Churn'] = df['Churn'].replace({'Yes': True, 'No': False})
```

```python
# Criando a coluna Contas_Diarias que proporciona uma visÃ£o mais detalhada do comportamento dos clientes ao longo do tempo
df['Contas_Diarias'] = df['Charges.Monthly'] / df['tenure']
df
```

```python
df.info()
```

#ðŸ“Š Carga e anÃ¡lise

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Contagem de clientes que saÃ­ram (True) e permaneceram (False)
churn_counts = df['Churn'].value_counts()

# Mapeando os rÃ³tulos
labels = ['Permanece na empresa', 'Saiu da empresa']
churn_counts.index = labels

# Definindo as cores
cores = ['olive', 'brown']

# Criando a figura
fig, ax = plt.subplots(figsize=(10, 8))

# Criando o grÃ¡fico de barras a partir das contagens
ax.bar(churn_counts.index, churn_counts.values, color=cores)

# Personalizando o grÃ¡fico
ax.set_title('DistribuiÃ§Ã£o da EvasÃ£o de Clientes', loc='left', fontsize=18)
ax.set_ylabel('Qtde. de Clientes', fontsize=14)
ax.set_xlabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12, rotation=0)

# Adicionando rÃ³tulos com a quantidade de clientes
for i, value in enumerate(churn_counts.values):
    ax.text(i, value + 50, str(value), ha='center', va='bottom', fontsize=12)

# Exibindo o grÃ¡fico
plt.show()
```

## Contagem de evasÃ£o por variÃ¡veis categÃ³ricas

```python
# Por gÃªnero
# Contagem de clientes por gÃªnero que saÃ­ram (True) e permaneceram (False)
churn_gender = df.groupby('gender')['Churn'].value_counts().unstack()

# Criando a figura
fig, ax = plt.subplots(figsize=(8, 8))

# Plotando as barras
churn_gender.plot(kind='bar', ax=ax, rot=0, color=['olive', 'brown'])

# Personalizando o grÃ¡fico
ax.set_title('EvasÃ£o de Clientes por GÃªnero', loc='left', fontsize=18)
ax.set_xlabel('GÃªnero', fontsize=14)
ax.set_ylabel('Qtde. de Clientes', fontsize=14)

# Ajustando a legenda
ax.legend(['Permaneceu', 'Saiu'], title='Status de Churn')

# Adicionando rÃ³tulos de valores em cada barra
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points',
                fontsize=12)

# Exibindo o grÃ¡fico
plt.show()
```

```python
# Por tipo de contrato
# Contagem de clientes por contrato que saÃ­ram (True) e permaneceram (False)
churn_contract = df.groupby('Contract')['Churn'].value_counts().unstack()

# Criando a figura
fig, ax = plt.subplots(figsize=(8, 8))

# Plotando as barras
churn_contract.plot(kind='bar', ax=ax, rot=0, color=['olive', 'brown'])

# Personalizando o grÃ¡fico
ax.set_title('EvasÃ£o de Clientes por Tipo de Contrato', loc='left', fontsize=18)
ax.set_xlabel('Tipo de Contrato', fontsize=14)
ax.set_ylabel('Qtde. de Clientes', fontsize=14)

# Ajustando a legenda
ax.legend(['Permaneceu', 'Saiu'], title='Status de Churn')

# Adicionando rÃ³tulos de valores em cada barra
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points',
                fontsize=12)

# Exibindo o grÃ¡fico
plt.show()
```

```python
# Por mÃ©todo de pagamento
# Contagem de clientes por mÃ©todo de pagamento que saÃ­ram (True) e permaneceram (False)
churn_payment = df.groupby('PaymentMethod')['Churn'].value_counts().unstack()

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 8))

# Plotando as barras
churn_payment.plot(kind='barh', ax=ax, rot=0, color=['olive', 'brown'])

# Personalizando o grÃ¡fico
ax.set_title('EvasÃ£o de Clientes por MÃ©todo de Pagamento', loc='left', fontsize=18)
ax.set_ylabel('MÃ©todo de Pagamento', fontsize=14)
ax.set_xlabel('Qtde. de Clientes', fontsize=14)

# Ajustando a legenda
ax.legend(['Permaneceu', 'Saiu'], title='Status de Churn')

# Adicionando rÃ³tulos de valores em cada barra
for p in ax.patches:
    # A posiÃ§Ã£o x (horizontal) Ã© a largura da barra
    x_val = p.get_width()
    # A posiÃ§Ã£o y (vertical) Ã© o centro da barra
    y_val = p.get_y() + p.get_height() / 2
    # Anotar o valor da largura na posiÃ§Ã£o correta
    ax.annotate(f'{int(x_val)}', (x_val, y_val), va='center', xytext=(5, 0), textcoords='offset points')


# Exibindo o grÃ¡fico
plt.show()
```
